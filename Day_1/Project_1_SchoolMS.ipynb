```python
course_catalogue = [
    {"course_id":1,"course_name":"python","credit":4},
    {"course_id":2,"course_name":"javascript","credit":3}
]

student_records = [
    {"name":"seud","email":"seudsahm1@gmail.com","enrolled_courses": [1]}
]
```

```python
def add_course(course_id, course_name, credit):
    # Create a new course dictionary
    new_course = {
        "course_id": course_id,
        "course_name": course_name,
        "credit": credit
    }
    # Add the new course to the course catalogue
    if any(course["course_id"] == course_id for course in course_catalogue):
        print(f"Course ID {course_id} already exists. Please choose a different ID.")
        return
    course_catalogue.append(new_course)
    print(f"Course '{course_name}' added successfully.")
```

```python
def view_courses():
    for course in course_catalogue:
        print(f'Course ID: {course["course_id"]}, Name: {course["course_name"]}, Credit: {course["credit"]}')
        
```

```python
def enroll_student(name, email, course_id):
    # Check if the course exists
    course_exists = any(course["course_id"] == course_id for course in course_catalogue)
    if not course_exists:
        print(f"Course ID {course_id} does not exist. Please choose a valid course.")
        return
    # Check if the student already exists
    student = next((student for student in student_records if student["email"] == email), None)
    if student:
        # If the student exists, add the course to their enrolled courses if not already enrolled
        if course_id not in student["enrolled_courses"]:
            student["enrolled_courses"].append(course_id)
            print(f"{name} has been enrolled in course ID {course_id}.")
        else:
            print(f"{name} is already enrolled in course ID {course_id}.")
    
```

```python
def update_student_email(old_email, new_email):
    student = next((student for student in student_records if student["email"] == old_email),None)
    if student:
        student["email"] = new_email
        print(f"Email updated successfully for {student['name']}.")
    else:
        print(f"No student found with email {old_email}.")
    
```

```python
def deregister_student(email, course_id):
    student = next((student for student in student_records if student["email"] == email), None)
    if student:
        if course_id in student["enrolled_courses"]:
            student["enrolled_courses"].remove(course_id)
            print(f"{student['name']} has been deregistered from course ID {course_id}.")
        else:
            print(f"{student['name']} is not enrolled in course ID {course_id}.")
    else:
        print(f"No student found with email {email}.")
```

```python
def enroll_course(email, course_id):
    student  = next((student for student in student_records if student["email"] == email), None)
    if student:
        if course_id not in student["enrolled_courses"]:
            student["enrolled_courses"].append(course_id)
            print(f"{student['name']} has been enrolled in course ID {course_id}.")
        else:
            print(f"{student['name']} is already enrolled in course ID {course_id}.")
    else:
        print(f"No student found with email {email}.")
```

```python
def drop_course(email, course_id):
    student = next((student for student in student_records if student["email"] == email), None)
    if student:
        if course_id in student["enrolled_courses"]:
            student["enrolled_courses"].remove(course_id)
            print(f"{student['name']} has been dropped from course ID {course_id}.")
        else:
            print(f"{student['name']} is not enrolled in course ID {course_id}.")
    else:
        print(f"No student found with email {email}.")
```
